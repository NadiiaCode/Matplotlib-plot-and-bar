```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file
gdp_by_country = pd.read_csv("data.csv")
#print csv file
print(gdp_by_country)
```

```python
# In Region column display countries located in Americas
gdp_by_country.loc[gdp_by_country["Region"] == "Americas"]
```

```python
# In Region column display countries located in Asia
gdp_by_country.loc[gdp_by_country["Region"] == "Asia"]
```

```python
# In Region column display countries located in Europe
gdp_by_country.loc[gdp_by_country["Region"] == "Europe"]
```

```python
#calculate the total GDP of 10 countries (Without USA and China)
gdp_estimate = gdp_by_country['Estimate'].sum()
print("Estimated GDP of top 10 countries (without USA and China):", gdp_estimate)
```

```python
#Create plot 
# Display 2 columns: x axis = Country, y axis = Estimate
columns = ["Country", "Estimate"]
plt.plot(gdp_by_country.Country, gdp_by_country.Estimate)
plt.show()
```

```python
columns = ["Country", "Estimate"]
#plot with 5 countries only
plt.plot(gdp_by_country.Country[:5], gdp_by_country.Estimate[:5])
#xticks with rotation 90 will display country names vertically
plt.xticks(rotation=90)
#added x and y labels
plt.xlabel("Country")
plt.ylabel("GDP in trillions (USD)")
#added title
plt.title("GDP per country")
plt.show()
```

```python
columns = ["Country", "Estimate"]
#added line color, linestyle, marker, linewidth, markerfacecolor, markersize
plt.plot(gdp_by_country.Country, gdp_by_country.Estimate, color="#dd0900",
         linestyle="--", marker="^", linewidth=2,  markerfacecolor='#007700', markersize=15)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("GDP in trillions (USD)")
plt.title("GDP per country")
#added grid
plt.grid(True)
plt.show()
```

```python
columns = ["Country", "Estimate"]
#changed line color, linestyle, marker, linewidth, markerfacecolor, markersize
plt.plot(gdp_by_country.Country, gdp_by_country.Estimate, color="#000000",
         linestyle="-.", marker="o", linewidth=3,  markerfacecolor='#007790', markersize=10)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("GDP in trillions (USD)")
plt.title("GDP per country")
plt.grid(True)
plt.show()
```

```python
# return to default style
plt.style.use("default")
```

```python
# create simple bar
columns = ["Country", "Estimate"]
plt.bar(gdp_by_country.Country, gdp_by_country.Estimate)
plt.show()
```

```python
columns = ["Country", "Estimate"]
# added bars color
plt.bar(gdp_by_country.Country, gdp_by_country.Estimate, color="#447878")
# add labels
plt.xlabel("Country")
plt.ylabel("GDP in trillions (USD)")
# change title text color, size and weight
plt.title("GDP per country", size=15, weight="bold", color="#444444")
plt.xticks(rotation=90)
# added tight layout
plt.tight_layout()
plt.show()
```

```python
columns = ["Country", "Estimate"]
# added edgecolor
bars = plt.bar(gdp_by_country.Country, gdp_by_country.Estimate, color="#447878",
               edgecolor="#ffff00")
# added set_hatch (all possible options just for the sake of displaying versatility)
bars[0].set_hatch("/")
bars[1].set_hatch("|")
bars[2].set_hatch("\\")
bars[3].set_hatch("X")
bars[4].set_hatch("-")
bars[5].set_hatch("+")
bars[6].set_hatch("*")
bars[7].set_hatch(".")
bars[8].set_hatch("o")
bars[9].set_hatch("O")
# added labels and title colors
plt.xlabel("Country", color="#444444")
plt.ylabel("GDP in trillions (USD)", color="#444444")
plt.title("GDP per country", size=15, weight="bold", color="#444444")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.style.use("default")
columns = ["Country", "Estimate"]
bars = plt.bar(gdp_by_country.Country, gdp_by_country.Estimate, color="#447878",
               edgecolor="#ffff00")
# set_hatch could be displayed with a combination of symbols
bars[0].set_hatch("/+")
bars[1].set_hatch("|-")
bars[2].set_hatch("\\|")
bars[3].set_hatch("**")
bars[4].set_hatch("..")
bars[5].set_hatch("-.")
bars[6].set_hatch("-*")
bars[7].set_hatch("./")
bars[8].set_hatch("oO")
bars[9].set_hatch("oo")
plt.xlabel("Country", color="#444444")
plt.ylabel("GDP in trillions (USD)", color="#444444")
plt.title("GDP per country", size=15, weight="bold", color="#444444")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
